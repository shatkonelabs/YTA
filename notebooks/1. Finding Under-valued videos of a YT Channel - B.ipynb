```python
channel = "freecodecamp"
%run ../scripts/1_channel_analysis.py
```

```python
vid = VideoInfoDownloader(channel).run()
raw_path = vid.video_info_path; raw_path
```

```python
cvr = ChannelVideosRanker(raw_path).run()
df = cvr.find_undervalued_videos()
```

```python
df.head(10)
```

```python
cvr.scatter_view(label="score")
```

```python
df[df['duration'] < 3600].head(10)
```

```python
for i, r in df.sort_values(by='score', ascending=False).head(10).iterrows():
    print(round(r['score'], 4), r['title'])
    print(f"https://www.youtube.com/watch?v={r['id']}", r['view_count'])
    print()
```
